```python
import  pandas as pd
import  numpy  as np
```

```python
df=pd.read_csv("/content/data.csv", encoding="cp932")
```

```python
import statsmodels.formula.api as smf
```

```python
bunseki =  "期末テスト ~ 小テスト"
```

```python
model= smf.ols(formula=bunseki,data=df).fit()
```

```python
model.summary()
```

```python
df["ln_期末"] = np.log1p(df["期末テスト"])  # ln(1 + x)
df["ln_小"]   = np.log1p(df["小テスト"])

model = smf.ols("ln_期末 ~ ln_小", data=df).fit()
print(model.summary())
```
