```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def minkowskiDist(v1, v2, p):
    dist = 0.0
    for i in range(len(v1)):
        dist += abs(v1[i] - v2[i])**p
    return dist**(1/p)

def euclidDist(f1, f2):
    return minkowskiDist(f1, f2, 2)


def compareAnimals(animals, precision=2, savepath=None):
    """
    animals: [(name, features), ...]
      name: str
      features: list/np.array (同じ長さ)
    """
    names = [name for name, _ in animals]
    feats = [np.array(features, dtype=float) for _, features in animals]

    tableVals = []
    for i in range(len(animals)):
        row = []
        for j in range(len(animals)):
            if i == j:
                row.append("--")
            else:
                d = euclidDist(feats[i], feats[j])
                row.append(str(round(d, precision)))
        tableVals.append(row)

    fig, ax = plt.subplots(figsize=(1 + 1.2*len(animals), 1 + 0.6*len(animals)))
    ax.axis("off")
    table = ax.table(
        rowLabels=names,
        colLabels=names,
        cellText=tableVals,
        cellLoc="center",
        loc="center",
    )
    table.scale(1, 2.0)
    if savepath:
        plt.savefig(savepath, dpi=200, bbox_inches="tight")
    plt.show()
```

```python
animals = [
    ("Dog",   [1, 0, 1, 0]),
    ("Cat",   [1, 0, 0.8, 0.1]),
    ("Horse", [0, 1, 0, 1]),
]

compareAnimals(animals, precision=2, savepath="distances.png")
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import pdist

def minkowskiDist(v1, v2, p):
    dist = 0.0
    for i in range(len(v1)):
        dist += abs(v1[i] - v2[i])**p
    return dist**(1/p)

def euclidDist(f1, f2):
    return minkowskiDist(f1, f2, 2)

# あなたのデータ
animals = [
    ("Dog",   [1, 0, 1, 0]),
    ("Cat",   [1, 0, 0.8, 0.1]),
    ("Horse", [0, 1, 0, 1]),
]

names = [n for n,_ in animals]
X = np.array([feat for _, feat in animals], dtype=float)

# 各ペア距離（ユークリッド）を作る（階層クラスタはこの形式が好き）
D = pdist(X, metric='euclidean')

# くっつけ方：'average'（平均距離） ほか 'single' 'complete' など
Z = linkage(D, method='average')

# デンドログラム表示
plt.figure(figsize=(6, 3))
dendrogram(Z, labels=names)
plt.title("Hierarchical Clustering (dendrogram)")
plt.show()

# 2クラスタに切る（例）
labels = fcluster(Z, t=2, criterion='maxclust')
print("cluster labels:", dict(zip(names, labels)))
```

あなたの距離表だとどうなる？

Dog–Cat = 0.22（近い）→ 同じクラスタになりやすい

Dog–Horse = 2.0（遠い）

Cat–Horse = 1.86（遠い）→ Horse は別クラスタになりやすい

つまり DogとCatが同じ組、Horseは別 という分け方が自然になります。

重要ポイント（これだけ覚えればOK）

クラスタリングは 「距離（似てる度）＋どこで切るか」 で決まります。
そして距離は、結局 特徴量をどう作るかで大きく変わります。
